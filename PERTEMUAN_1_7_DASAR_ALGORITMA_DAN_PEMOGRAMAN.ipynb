```python
import math

def hitung_keliling_lingkaran(jari_jari):
    keliling = 2 * math.pi * jari_jari
    return keliling

def main():
    print("Kalkulator Keliling Lingkaran")

    try:
        jari_jari = float(input("Masukkan jari-jari lingkaran: "))

        if jari_jari < 0:
            print("Jari-jari tidak boleh negatif.")
        else:
            keliling = hitung_keliling_lingkaran(jari_jari)
            print(f"Keliling lingkaran dengan jari-jari {jari_jari} adalah: {keliling:.2f}")

    except ValueError:
        print("Input tidak valid. Harap masukkan angka.")

if __name__ == "__main__":
    main()
```

```python
import math

def hitung_luas_segitiga(a, b, c):
    # Menghitung setengah keliling
    s = (a + b + c) / 2
    # Menghitung luas menggunakan rumus Heron
    luas = math.sqrt(s * (s - a) * (s - b) * (s - c))
    return luas

def hitung_keliling_segitiga(a, b, c):
    # Menghitung keliling segitiga
    keliling = a + b + c
    return keliling

def main():
    print("Kalkulator Luas dan Keliling Segitiga (Rumus Heron)")

    try:
        # Meminta input dari pengguna
        a = float(input("Masukkan panjang sisi a: "))
        b = float(input("Masukkan panjang sisi b: "))
        c = float(input("Masukkan panjang sisi c: "))

        # Memeriksa apakah sisi-sisi membentuk segitiga
        if a + b > c and a + c > b and b + c > a:
            luas = hitung_luas_segitiga(a, b, c)
            keliling = hitung_keliling_segitiga(a, b, c)
            print(f"Luas segitiga: {luas:.2f}")
            print(f"Keliling segitiga: {keliling:.2f}")
        else:
            print("Sisi-sisi yang dimasukkan tidak membentuk segitiga.")

    except ValueError:
        print("Input tidak valid. Harap masukkan angka.")

if __name__ == "__main__":
    main()
```

```python
Sdef tambah(x, y):
    return x + y

def kurang(x, y):
    return x - y

def kali(x, y):
    return x * y

def bagi(x, y):
    if y == 0:
        return "Error: Pembagian dengan nol tidak diperbolehkan."
    return x / y

def main():
    print("Kalkulator Sederhana")
    print("Pilih operasi:")
    print("1. Tambah")
    print("2. Kurang")
    print("3. Kali")
    print("4. Bagi")

    pilihan = input("Masukkan pilihan (1/2/3/4): ")

    if pilihan in ['1', '2', '3', '4']:
        try:
            angka1 = float(input("Masukkan angka pertama: "))
            angka2 = float(input("Masukkan angka kedua: "))

            if pilihan == '1':
                print(f"{angka1} + {angka2} = {tambah(angka1, angka2)}")
            elif pilihan == '2':
                print(f"{angka1} - {angka2} = {kurang(angka1, angka2)}")
            elif pilihan == '3':
                print(f"{angka1} * {angka2} = {kali(angka1, angka2)}")
            elif pilihan == '4':
                print(f"{angka1} / {angka2} = {bagi(angka1, angka2)}")
        except ValueError:
            print("Input tidak valid. Harap masukkan angka.")
    else:
        print("Pilihan tidak valid!")

if __name__ == "__main__":
    main()
```

```python
def konversi_suhu():
    print("=== Program Konversi Suhu ===")
    print("Pilih jenis konversi:")
    print("1. Celsius ke Fahrenheit")
    print("2. Celsius ke Kelvin")
    print("3. Fahrenheit ke Celsius")
    print("4. Fahrenheit ke Kelvin")
    print("5. Kelvin ke Celsius")
    print("6. Kelvin ke Fahrenheit")

    pilihan = input("Masukkan pilihan (1-6): ")

    if pilihan == '1':
        celsius = float(input("Masukkan suhu dalam Celsius: "))
        fahrenheit = (celsius * 9/5) + 32
        print(f"{celsius} °C = {fahrenheit} °F")

    elif pilihan == '2':
        celsius = float(input("Masukkan suhu dalam Celsius: "))
        kelvin = celsius + 273.15
        print(f"{celsius} °C = {kelvin} K")

    elif pilihan == '3':
        fahrenheit = float(input("Masukkan suhu dalam Fahrenheit: "))
        celsius = (fahrenheit - 32) * 5/9
        print(f"{fahrenheit} °F = {celsius} °C")

    elif pilihan == '4':
        fahrenheit = float(input("Masukkan suhu dalam Fahrenheit: "))
        kelvin = (fahrenheit - 32) * 5/9 + 273.15
        print(f"{fahrenheit} °F = {kelvin} K")

    elif pilihan == '5':
        kelvin = float(input("Masukkan suhu dalam Kelvin: "))
        celsius = kelvin - 273.15
        print(f"{kelvin} K = {celsius} °C")

    elif pilihan == '6':
        kelvin = float(input("Masukkan suhu dalam Kelvin: "))
        fahrenheit = (kelvin - 273.15) * 9/5 + 32
        print(f"{kelvin} K = {fahrenheit} °F")

    else:
        print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    konversi_suhu()
```

```python
# Sistem Informasi Akademik Sederhana

mahasiswa = {}
mata_kuliah = {}
nilai = {}

def tambah_mahasiswa():
    nim = input("Masukkan NIM: ")
    nama = input("Masukkan Nama Mahasiswa: ")
    mahasiswa[nim] = {"nama": nama}
    print(f"Mahasiswa {nama} berhasil ditambahkan.\n")

def tambah_mata_kuliah():
    kode = input("Masukkan Kode Mata Kuliah: ")
    nama = input("Masukkan Nama Mata Kuliah: ")
    sks = int(input("Masukkan Jumlah SKS: "))
    mata_kuliah[kode] = {"nama": nama, "sks": sks}
    print(f"Mata kuliah {nama} berhasil ditambahkan.\n")

def input_nilai():
    nim = input("Masukkan NIM Mahasiswa: ")
    kode = input("Masukkan Kode Mata Kuliah: ")
    if nim in mahasiswa and kode in mata_kuliah:
        nilai_mhs = float(input("Masukkan Nilai (0-100): "))
        nilai.setdefault(nim, {})[kode] = nilai_mhs
        print("Nilai berhasil disimpan.\n")
    else:
        print("NIM atau Kode Mata Kuliah tidak ditemukan.\n")

def tampilkan_transkrip():
    nim = input("Masukkan NIM Mahasiswa: ")
    if nim in mahasiswa:
        print(f"\nTranskrip Nilai - {mahasiswa[nim]['nama']} ({nim})")
        total_sks = 0
        total_bobot = 0
        for kode, nilai_mhs in nilai.get(nim, {}).items():
            mk = mata_kuliah[kode]
            bobot = hitung_bobot(nilai_mhs)
            print(f"{mk['nama']} ({mk['sks']} SKS): {nilai_mhs} => {bobot}")
            total_sks += mk['sks']
            total_bobot += bobot * mk['sks']
        if total_sks > 0:
            ipk = total_bobot / total_sks
            print(f"IPK: {ipk:.2f}\n")
        else:
            print("Belum ada nilai.\n")
    else:
        print("Mahasiswa tidak ditemukan.\n")

def hitung_bobot(nilai):
    if nilai >= 85:
        return 4.0
    elif nilai >= 75:
        return 3.0
    elif nilai >= 65:
        return 2.0
    elif nilai >= 50:
        return 1.0
    else:
        return 0.0

def menu():
    while True:
        print("=== Sistem Informasi Akademik ===")
        print("1. Tambah Mahasiswa")
        print("2. Tambah Mata Kuliah")
        print("3. Input Nilai")
        print("4. Tampilkan Transkrip")
        print("5. Keluar")
        pilihan = input("Pilih menu (1-5): ")

        if pilihan == '1':
            tambah_mahasiswa()
        elif pilihan == '2':
            tambah_mata_kuliah()
        elif pilihan == '3':
            input_nilai()
        elif pilihan == '4':
            tampilkan_transkrip()
        elif pilihan == '5':
            print("Terima kasih telah menggunakan sistem ini.")
            break
        else:
            print("Pilihan tidak valid.\n")

if __name__ == "__main__":
    menu()
```

```python
# Aplikasi E-commerce Sederhana (Command Line)

produk = {
    "P001": {"nama": "Kaos Polos", "harga": 50000},
    "P002": {"nama": "Celana Jeans", "harga": 150000},
    "P003": {"nama": "Sneakers", "harga": 300000},
}

keranjang = {}

def tampilkan_produk():
    print("\n=== Daftar Produk ===")
    for kode, info in produk.items():
        print(f"{kode}: {info['nama']} - Rp {info['harga']:,}")
    print()

def tambah_ke_keranjang():
    kode = input("Masukkan kode produk: ").upper()
    if kode in produk:
        jumlah = int(input("Masukkan jumlah: "))
        if kode in keranjang:
            keranjang[kode] += jumlah
        else:
            keranjang[kode] = jumlah
        print("Produk berhasil ditambahkan ke keranjang.\n")
    else:
        print("Kode produk tidak ditemukan.\n")

def tampilkan_keranjang():
    if not keranjang:
        print("Keranjang masih kosong.\n")
        return

    print("\n=== Keranjang Belanja ===")
    total = 0
    for kode, jumlah in keranjang.items():
        item = produk[kode]
        subtotal = item["harga"] * jumlah
        total += subtotal
        print(f"{item['nama']} x{jumlah} - Rp {subtotal:,}")
    print(f"Total Belanja: Rp {total:,}\n")

def checkout():
    if not keranjang:
        print("Keranjang kosong, tidak bisa checkout.\n")
        return

    tampilkan_keranjang()
    konfirmasi = input("Lanjutkan checkout? (y/n): ").lower()
    if konfirmasi == 'y':
        simpan_transaksi()
        keranjang.clear()
        print("Transaksi berhasil! Terima kasih sudah belanja.\n")
    else:
        print("Checkout dibatalkan.\n")

def simpan_transaksi():
    with open("transaksi.txt", "a") as file:
        file.write("=== Transaksi Baru ===\n")
        total = 0
        for kode, jumlah in keranjang.items():
            item = produk[kode]
            subtotal = item["harga"] * jumlah
            total += subtotal
            file.write(f"{item['nama']} x{jumlah} - Rp {subtotal:,}\n")
        file.write(f"Total: Rp {total:,}\n\n")

def menu():
    while True:
        print("=== Aplikasi E-Commerce Sederhana ===")
        print("1. Lihat Produk")
        print("2. Tambah ke Keranjang")
        print("3. Lihat Keranjang")
        print("4. Checkout")
        print("5. Keluar")

        pilihan = input("Pilih menu (1-5): ")

        if pilihan == '1':
            tampilkan_produk()
        elif pilihan == '2':
            tambah_ke_keranjang()
        elif pilihan == '3':
            tampilkan_keranjang()
        elif pilihan == '4':
            checkout()
        elif pilihan == '5':
            print("Terima kasih sudah menggunakan aplikasi ini.")
            break
        else:
            print("Pilihan tidak valid.\n")

if __name__ == "__main__":
    menu()
```

```python
# Sistem Reservasi Hotel Sederhana

kamar = {
    "101": {"tipe": "Single", "harga": 300000, "status": "kosong"},
    "102": {"tipe": "Single", "harga": 300000, "status": "kosong"},
    "201": {"tipe": "Double", "harga": 500000, "status": "kosong"},
    "202": {"tipe": "Double", "harga": 500000, "status": "kosong"},
}

reservasi = {}

def tampilkan_kamar():
    print("\n=== Daftar Kamar ===")
    for no, info in kamar.items():
        print(f"Kamar {no}: {info['tipe']} - Rp {info['harga']:,} - {info['status']}")
    print()

def reservasi_kamar():
    nama = input("Masukkan nama tamu: ")
    no_kamar = input("Masukkan nomor kamar yang ingin dipesan: ")

    if no_kamar in kamar:
        if kamar[no_kamar]["status"] == "kosong":
            durasi = int(input("Masukkan lama menginap (hari): "))
            total = kamar[no_kamar]["harga"] * durasi
            kamar[no_kamar]["status"] = "dipesan"
            reservasi[no_kamar] = {
                "nama": nama,
                "durasi": durasi,
                "total": total
            }
            print(f"Reservasi berhasil. Total bayar: Rp {total:,}\n")
        else:
            print("Maaf, kamar sudah dipesan.\n")
    else:
        print("Nomor kamar tidak ditemukan.\n")

def tampilkan_reservasi():
    if not reservasi:
        print("Belum ada reservasi.\n")
        return

    print("\n=== Daftar Reservasi ===")
    for no_kamar, data in reservasi.items():
        print(f"Kamar {no_kamar} - {data['nama']} - {data['durasi']} hari - Total: Rp {data['total']:,}")
    print()

def checkout():
    no_kamar = input("Masukkan nomor kamar untuk checkout: ")
    if no_kamar in reservasi:
        kamar[no_kamar]["status"] = "kosong"
        del reservasi[no_kamar]
        print("Checkout berhasil. Kamar tersedia kembali.\n")
    else:
        print("Tidak ada reservasi untuk kamar tersebut.\n")

def menu():
    while True:
        print("=== Sistem Reservasi Hotel ===")
        print("1. Lihat Daftar Kamar")
        print("2. Reservasi Kamar")
        print("3. Lihat Reservasi")
        print("4. Checkout")
        print("5. Keluar")
        pilihan = input("Pilih menu (1-5): ")

        if pilihan == '1':
            tampilkan_kamar()
        elif pilihan == '2':
            reservasi_kamar()
        elif pilihan == '3':
            tampilkan_reservasi()
        elif pilihan == '4':
            checkout()
        elif pilihan == '5':
            print("Terima kasih telah menggunakan sistem reservasi hotel.")
            break
        else:
            print("Pilihan tidak valid.\n")

if __name__ == "__main__":
    menu()
```

```python
def hitung_nilai_akhir(tugas, uts, uas):
    return (0.3 * tugas) + (0.3 * uts) + (0.4 * uas)

def tentukan_grade(nilai_akhir):
    if nilai_akhir >= 85:
        return "A", "Sangat Baik / Lulus"
    elif nilai_akhir >= 75:
        return "B", "Baik / Lulus"
    elif nilai_akhir >= 65:
        return "C", "Cukup / Lulus"
    elif nilai_akhir >= 50:
        return "D", "Kurang / Tidak Lulus"
    else:
        return "E", "Sangat Kurang / Tidak Lulus"

def main():
    print("=== Program Penilaian Mahasiswa ===")
    nama = input("Masukkan nama mahasiswa: ")
    nim = input("Masukkan NIM mahasiswa: ")

    try:
        tugas = float(input("Masukkan nilai Tugas (0-100): "))
        uts = float(input("Masukkan nilai UTS (0-100): "))
        uas = float(input("Masukkan nilai UAS (0-100): "))

        nilai_akhir = hitung_nilai_akhir(tugas, uts, uas)
        grade, keterangan = tentukan_grade(nilai_akhir)

        print("\n=== Hasil Penilaian ===")
        print(f"Nama       : {nama}")
        print(f"NIM        : {nim}")
        print(f"Nilai Akhir: {nilai_akhir:.2f}")
        print(f"Grade      : {grade}")
        print(f"Keterangan : {keterangan}\n")
    except ValueError:
        print("Input harus berupa angka!")

if __name__ == "__main__":
    main()
```

```python
# Simulasi ATM Sederhana

# Data awal (akun contoh)
akun = {
    'nomor_rekening': '1234567890',
    'pin': '301105',
    'saldo': 1000000  # saldo awal dalam rupiah
}

# Akun tujuan transfer (dummy)
akun_tujuan = {
    '9876543210': 500000
}

def verifikasi_pin():
    percobaan = 3
    while percobaan > 0:
        pin = input("Masukkan PIN Anda: ")
        if pin == akun['pin']:
            print("PIN benar. Selamat datang!\n")
            return True
        else:
            percobaan -= 1
            print(f"PIN salah. Sisa percobaan: {percobaan}")
    print("Akun terblokir. Silakan hubungi bank.")
    return False

def cek_saldo():
    print(f"Saldo Anda saat ini: Rp{akun['saldo']:,}\n")

def tarik_tunai():
    try:
        jumlah = int(input("Masukkan jumlah yang ingin ditarik: Rp"))
        if jumlah <= akun['saldo']:
            akun['saldo'] -= jumlah
            print(f"Penarikan berhasil. Sisa saldo: Rp{akun['saldo']:,}\n")
        else:
            print("Saldo tidak mencukupi.\n")
    except ValueError:
        print("Masukkan jumlah yang valid.\n")

def setor_tunai():
    try:
        jumlah = int(input("Masukkan jumlah yang ingin disetor: Rp"))
        akun['saldo'] += jumlah
        print(f"Setoran berhasil. Saldo Anda sekarang: Rp{akun['saldo']:,}\n")
    except ValueError:
        print("Masukkan jumlah yang valid.\n")

def transfer():
    tujuan = input("Masukkan nomor rekening tujuan: ")
    if tujuan not in akun_tujuan:
        print("Nomor rekening tidak ditemukan.\n")
        return
    try:
        jumlah = int(input("Masukkan jumlah yang ingin ditransfer: Rp"))
        if jumlah <= akun['saldo']:
            akun['saldo'] -= jumlah
            akun_tujuan[tujuan] += jumlah
            print(f"Transfer berhasil ke {tujuan}. Sisa saldo: Rp{akun['saldo']:,}\n")
        else:
            print("Saldo tidak mencukupi.\n")
    except ValueError:
        print("Masukkan jumlah yang valid.\n")

def ganti_pin():
    pin_lama = input("Masukkan PIN lama: ")
    if pin_lama == akun['pin']:
        pin_baru = input("Masukkan PIN baru: ")
        konfirmasi = input("Konfirmasi PIN baru: ")
        if pin_baru == konfirmasi:
            akun['pin'] = pin_baru
            print("PIN berhasil diganti.\n")
        else:
            print("Konfirmasi PIN tidak cocok.\n")
    else:
        print("PIN lama salah.\n")

def menu():
    while True:
        print("===== Menu ATM =====")
        print("1. Cek Saldo")
        print("2. Tarik Tunai")
        print("3. Setor Tunai")
        print("4. Transfer")
        print("5. Ganti PIN")
        print("6. Keluar")
        pilihan = input("Pilih menu (1-6): ")
        print()
        if pilihan == '1':
            cek_saldo()
        elif pilihan == '2':
            tarik_tunai()
        elif pilihan == '3':
            setor_tunai()
        elif pilihan == '4':
            transfer()
        elif pilihan == '5':
            ganti_pin()
        elif pilihan == '6':
            print("Terima kasih telah menggunakan layanan ATM kami.")
            break
        else:
            print("Pilihan tidak valid.\n")

# Main Program
if verifikasi_pin():
    menu()
```

```python
def fibonacci(n):
    fib_sequence = []
    a, b = 0, 1
    for _ in range(n):
        fib_sequence.append(a)
        a, b = b, a + b
    return fib_sequence

# Input dari pengguna
n = int(input("Masukkan jumlah angka Fibonacci yang ingin ditampilkan: "))
result = fibonacci(n)

print("Deret Fibonacci hingga", n, "angka:")
print(result)
```

```python
def add_matrices(A, B):
    return [[A[i][j] + B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def subtract_matrices(A, B):
    return [[A[i][j] - B[i][j] for j in range(len(A[0]))] for i in range(len(A))]

def multiply_matrices(A, B):
    return [[sum(A[i][k] * B[k][j] for k in range(len(B))) for j in range(len(B[0]))] for i in range(len(A))]

# Contoh penggunaan
A = [[1, 2, 3], [4, 5, 6]]
B = [[7, 8, 9], [10, 11, 12]]
C = [[1, 2], [3, 4], [5, 6]]

print("A + B =", add_matrices(A, B))
print("A - B =", subtract_matrices(A, B))
print("A * C =", multiply_matrices(A, C))
```

```python
def cari_maks_min(array):
    if not array:
        return None, None  # Kalau array kosong

    maksimum = array[0]
    minimum = array[0]

    for angka in array:
        if angka > maksimum:
            maksimum = angka
        if angka < minimum:
            minimum = angka

    return maksimum, minimum

# Contoh penggunaan
data = [10, 45, 2, 33, 5, 99, -7, 0]

maks, min = cari_maks_min(data)

print(f"Array: {data}")
print(f"Nilai maksimum: {maks}")
print(f"Nilai minimum: {min}")
```

```python
def hitung_statistik(teks):
    jumlah_kata = len(teks.split())
    jumlah_karakter = len(teks)
    jumlah_kalimat = teks.count('.') + teks.count('!') + teks.count('?')
    return jumlah_kata, jumlah_karakter, jumlah_kalimat

def cari_kata(teks, kata):
    jumlah = teks.lower().count(kata.lower())
    return jumlah

def ganti_teks(teks, cari, ganti):
    return teks.replace(cari, ganti)

def format_teks(teks, mode):
    if mode == 'uppercase':
        return teks.upper()
    elif mode == 'lowercase':
        return teks.lower()
    elif mode == 'title':
        return teks.title()
    else:
        return teks

def main():
    teks = input("Masukkan teks: ")

    while True:
        print("\nMenu:")
        print("1. Hitung Statistik")
        print("2. Cari Kata")
        print("3. Ganti Teks")
        print("4. Format Teks")
        print("5. Tampilkan Teks")
        print("6. Keluar")

        pilihan = input("Pilih menu (1-6): ")

        if pilihan == '1':
            kata, karakter, kalimat = hitung_statistik(teks)
            print(f"Jumlah kata: {kata}")
            print(f"Jumlah karakter: {karakter}")
            print(f"Jumlah kalimat: {kalimat}")

        elif pilihan == '2':
            kata_dicari = input("Masukkan kata yang ingin dicari: ")
            jumlah = cari_kata(teks, kata_dicari)
            print(f"Kata '{kata_dicari}' ditemukan sebanyak {jumlah} kali.")

        elif pilihan == '3':
            cari = input("Masukkan kata yang ingin diganti: ")
            ganti = input("Masukkan kata pengganti: ")
            teks = ganti_teks(teks, cari, ganti)
            print("Teks berhasil diganti.")

        elif pilihan == '4':
            print("Pilihan format: uppercase / lowercase / title")
            mode = input("Masukkan format: ").lower()
            teks = format_teks(teks, mode)
            print("Format teks berhasil diubah.")

        elif pilihan == '5':
            print("\nTeks saat ini:")
            print(teks)

        elif pilihan == '6':
            print("Terima kasih telah menggunakan aplikasi pengolah teks!")
            break

        else:
            print("Pilihan tidak valid, silakan coba lagi.")

if __name__ == "__main__":
    main()
```

```python
def tambah(a, b):
    """
    Menjumlahkan dua angka.
    """
    return a + b

def kurang(a, b):
    """
    Mengurangi dua angka.
    """
    return a - b

def kali(a, b):
    """
    Mengalikan dua angka.
    """
    return a * b

def bagi(a, b):
    """
    Membagi dua angka.
    Mengembalikan hasil pembagian a / b.
    Jika b adalah 0, akan mengembalikan None dan mencetak peringatan.
    """
    if b == 0:
        print("Error: Pembagian dengan nol tidak diperbolehkan.")
        return None
    return a / b

def pangkat(a, b):
    """
    Menghitung a pangkat b.
    """
    return a ** b

def faktorial(n):
    """
    Menghitung faktorial dari n secara rekursif.
    Faktorial hanya didefinisikan untuk bilangan bulat non-negatif.
    """
    if not isinstance(n, int) or n < 0:
        raise ValueError("Faktorial hanya untuk bilangan bulat non-negatif.")
    if n == 0 or n == 1:
        return 1
    else:
        return n * faktorial(n - 1)

def rata_rata(data):
    """
    Menghitung rata-rata dari list data numerik.
    """
    if not data:
        return None
    return sum(data) / len(data)

def median(data):
    """
    Menghitung median dari list data numerik.
    """
    if not data:
        return None
    sorted_data = sorted(data)
    n = len(sorted_data)
    middle = n // 2
    if n % 2 == 1:
        return sorted_data[middle]
    else:
        return (sorted_data[middle - 1] + sorted_data[middle]) / 2

# Contoh penggunaan fungsi dan outputnya
if __name__ == "__main__":
    print("Tambah 5 + 3 =", tambah(5, 3))
    print("Kurang 5 - 3 =", kurang(5, 3))
    print("Kali 5 * 3 =", kali(5, 3))
    print("Bagi 5 / 3 =", bagi(5, 3))
    print("Pangkat 5^3 =", pangkat(5, 3))
    print("Faktorial 5! =", faktorial(5))
    data = [10, 20, 30, 40, 50]
    print("Rata-rata dari", data, "=", rata_rata(data))
    print("Median dari", data, "=", median(data))
```

```python
def tambah_data(mahasiswa_list):
    """
    Menambah data mahasiswa ke dalam list.
    """
    nim = input("Masukkan NIM mahasiswa: ").strip()
    if any(m['nim'] == nim for m in mahasiswa_list):
        print(f"NIM {nim} sudah ada dalam data.")
        return
    nama = input("Masukkan nama mahasiswa: ").strip()
    try:
        nilai = float(input("Masukkan nilai mahasiswa (0-100): "))
        if not (0 <= nilai <= 100):
            print("Nilai harus antara 0 sampai 100.")
            return
    except ValueError:
        print("Nilai harus berupa angka.")
        return
    mahasiswa_list.append({"nim": nim, "nama": nama, "nilai": nilai})
    print(f"Data mahasiswa {nama} berhasil ditambahkan.")

def cari_data(mahasiswa_list):
    """
    Mencari dan menampilkan data mahasiswa berdasarkan NIM atau nama.
    """
    keyword = input("Cari data mahasiswa berdasarkan NIM atau nama: ").strip().lower()
    hasil = [m for m in mahasiswa_list if keyword in m['nim'].lower() or keyword in m['nama'].lower()]
    if hasil:
        print(f"Ditemukan {len(hasil)} data:")
        for m in hasil:
            print(f"NIM: {m['nim']}, Nama: {m['nama']}, Nilai: {m['nilai']}")
    else:
        print("Data mahasiswa tidak ditemukan.")

def hitung_rata_rata(mahasiswa_list):
    """
    Menghitung rata-rata nilai semua mahasiswa.
    """
    if not mahasiswa_list:
        print("Data mahasiswa kosong.")
        return
    total_nilai = sum(m['nilai'] for m in mahasiswa_list)
    rata = total_nilai / len(mahasiswa_list)
    print(f"Rata-rata nilai semua mahasiswa adalah: {rata:.2f}")

def tampilkan_berdasarkan_peringkat(mahasiswa_list):
    """
    Menampilkan data mahasiswa berdasarkan peringkat nilai (tertinggi ke terendah).
    """
    if not mahasiswa_list:
        print("Data mahasiswa kosong.")
        return
    sorted_list = sorted(mahasiswa_list, key=lambda x: x['nilai'], reverse=True)
    print(f"{'Peringkat':<10} {'NIM':<15} {'Nama':<20} {'Nilai':<6}")
    print("-" * 55)
    for i, m in enumerate(sorted_list, start=1):
        print(f"{i:<10} {m['nim']:<15} {m['nama']:<20} {m['nilai']:<6.2f}")

def main():
    mahasiswa_list = []
    while True:
        print("\nMenu:")
        print("1. Tambah data mahasiswa")
        print("2. Cari data mahasiswa")
        print("3. Hitung rata-rata nilai mahasiswa")
        print("4. Tampilkan data mahasiswa berdasarkan peringkat nilai")
        print("5. Keluar")
        pilihan = input("Pilih menu (1-5): ").strip()
        if pilihan == '1':
            tambah_data(mahasiswa_list)
        elif pilihan == '2':
            cari_data(mahasiswa_list)
        elif pilihan == '3':
            hitung_rata_rata(mahasiswa_list)
        elif pilihan == '4':
            tampilkan_berdasarkan_peringkat(mahasiswa_list)
        elif pilihan == '5':
            print("Terima kasih!")
            break
        else:
            print("Pilihan tidak valid, silakan coba lagi.")

if __name__ == "__main__":
    main()
```

```python
import re
from datetime import datetime

# --- Validasi Input ---

def validasi_email(email):
    """
    Validasi format email sederhana.
    Mengembalikan True jika email valid, False jika tidak.
    """
    pola = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    return bool(re.match(pola, email))

def validasi_nomor_telepon(nomor):
    """
    Validasi nomor telepon Indonesia sederhana.
    Format diterima: angka 10-13 digit, boleh diawali dengan '+' dan kode negara.
    """
    pola = r'^(\+62|62|0)[2-9][0-9]{7,11}$'
    return bool(re.match(pola, nomor))

def validasi_tanggal(tanggal_str, format='%Y-%m-%d'):
    """
    Validasi apakah string tanggal sesuai format yang diberikan (default 'YYYY-MM-DD').
    Mengembalikan True jika valid, False jika tidak.
    """
    try:
        datetime.strptime(tanggal_str, format)
        return True
    except ValueError:
        return False

# --- Manipulasi String ---

def hitung_kata(teks):
    """
    Menghitung jumlah kata dalam string.
    """
    kata = re.findall(r'\b\w+\b', teks)
    return len(kata)

def ubah_format_teks(teks, format_tujuan):
    """
    Mengubah format teks.
    format_tujuan: 'upper' untuk huruf kapital seluruhnya,
                   'lower' untuk huruf kecil seluruhnya,
                   'title' untuk kapitalisasi judul.
    """
    if format_tujuan == 'upper':
        return teks.upper()
    elif format_tujuan == 'lower':
        return teks.lower()
    elif format_tujuan == 'title':
        return teks.title()
    else:
        raise ValueError("Format tujuan tidak dikenali. Gunakan 'upper', 'lower', atau 'title'.")

# --- Operasi File ---

def baca_file(path):
    """
    Membaca isi file teks dan mengembalikan sebagai string.
    """
    try:
        with open(path, 'r', encoding='utf-8') as f:
            return f.read()
    except FileNotFoundError:
        print(f"File {path} tidak ditemukan.")
        return None
    except IOError as e:
        print(f"Terjadi error saat membaca file: {e}")
        return None

def tulis_file(path, isi, mode='w'):
    """
    Menulis string isi ke file.
    mode default 'w' untuk overwrite, bisa diubah ke 'a' untuk append.
    """
    try:
        with open(path, mode, encoding='utf-8') as f:
            f.write(isi)
        print(f"Berhasil menulis ke file {path}.")
    except IOError as e:
        print(f"Terjadi error saat menulis file: {e}")

def proses_data_dari_file(path, fungsi_proses):
    """
    Membaca isi file, lalu memprosesnya dengan fungsi_proses yang diberikan.
    fungsi_proses adalah fungsi yang menerima satu argumen string isi file dan mengembalikan hasil proses.
    """
    isi = baca_file(path)
    if isi is not None:
        return fungsi_proses(isi)
    else:
        return None

# Contoh jika dijalankan langsung untuk test singkat
if __name__ == "__main__":
    # Test validasi
    print("Valid email 'zaidansujuda@gmail.com:", validasi_email('zaidansujuda@gmail.com'))
    print("Valid nomor telepon '+6281234567890':", validasi_nomor_telepon('+6281234567890'))
    print("Valid tanggal '2025-04-01':", validasi_tanggal('2025-04-01'))

    # Test manipulasi string
    kalimat = "Halo Untara!"
    print("Jumlah kata:", hitung_kata(kalimat))
    print("Uppercase:", ubah_format_teks(kalimat, 'upper'))
    print("Titlecase:", ubah_format_teks(kalimat, 'title'))

    # Test operasi file (pastikan ada file test.txt)
    isi_file = baca_file('test.txt')
    if isi_file:
        print("Isi file test.txt:")
        print(isi_file)

    tulis_file('output.txt', 'Ini contoh tulisan.\n')

    def hitung_panjang(teks):
        return f"Panjang teks: {len(teks)} karakter."

    hasil_proses = proses_data_dari_file('test.txt', hitung_panjang)
    if hasil_proses:
        print(hasil_proses)
```

```python
import time

# Faktorial implementations

def faktorial_rekursi(n):
    """Faktorial menggunakan rekursi dasar"""
    if n == 0 or n == 1:
        return 1
    return n * faktorial_rekursi(n - 1)

def faktorial_memo(n, memo={}):
    """Faktorial menggunakan rekursi dengan memoization"""
    if n in memo:
        return memo[n]
    if n == 0 or n == 1:
        memo[n] = 1
    else:
        memo[n] = n * faktorial_memo(n - 1, memo)
    return memo[n]

def faktorial_iterasi(n):
    """Faktorial menggunakan iterasi"""
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Fibonacci implementations

def fibonacci_rekursi(n):
    """Fibonacci menggunakan rekursi dasar"""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    return fibonacci_rekursi(n - 1) + fibonacci_rekursi(n - 2)

def fibonacci_memo(n, memo={}):
    """Fibonacci menggunakan rekursi dengan memoization"""
    if n in memo:
        return memo[n]
    if n <= 0:
        memo[n] = 0
    elif n == 1:
        memo[n] = 1
    else:
        memo[n] = fibonacci_memo(n - 1, memo) + fibonacci_memo(n - 2, memo)
    return memo[n]

def fibonacci_iterasi(n):
    """Fibonacci menggunakan iterasi"""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    a, b = 0, 1
    for _ in range(2, n + 1):
        a, b = b, a + b
    return b

# Benchmarking helper function

def benchmark(func, args):
    start = time.time()
    result = func(args)
    end = time.time()
    elapsed = end - start
    return elapsed, result

def test_faktorial():
    print("=== Benchmark Faktorial ===")
    test_values = [5, 10, 15, 20]
    for n in test_values:
        print(f"\nInput: {n}")
        t1, r1 = benchmark(faktorial_rekursi, n)
        print(f"Rekursi dasar: {t1:.8f} detik, hasil = {r1}")

        t2, r2 = benchmark(lambda x: faktorial_memo(x, {}), n)
        print(f"Rekursi memoization: {t2:.8f} detik, hasil = {r2}")

        t3, r3 = benchmark(faktorial_iterasi, n)
        print(f"Iterasi: {t3:.8f} detik, hasil = {r3}")

def test_fibonacci():
    print("\n=== Benchmark Fibonacci ===")
    test_values = [10, 20, 30, 35]
    for n in test_values:
        print(f"\nInput: {n}")
        t1, r1 = benchmark(fibonacci_rekursi, n)
        print(f"Rekursi dasar: {t1:.8f} detik, hasil = {r1}")

        t2, r2 = benchmark(lambda x: fibonacci_memo(x, {}), n)
        print(f"Rekursi memoization: {t2:.8f} detik, hasil = {r2}")

        t3, r3 = benchmark(fibonacci_iterasi, n)
        print(f"Iterasi: {t3:.8f} detik, hasil = {r3}")

def kesimpulan():
    print("""
KESIMPULAN:
- Rekursi dasar untuk faktorial dan fibonacci mudah dipahami dan diimplementasikan,
  tetapi untuk nilai input besar, khususnya fibonacci, sangat tidak efisien (waktu eksponensial).
- Rekursi dengan memoization secara signifikan mempercepat perhitungan karena menggunakan
  cache untuk menghindari perhitungan ulang.
- Pendekatan iteratif adalah yang paling efisien dalam penggunaan waktu dan memori untuk kedua kasus.
- Untuk aplikasi nyata, sebaiknya gunakan pendekatan memoization atau iterasi untuk performa terbaik.
""")

if __name__ == "__main__":
    test_faktorial()
    test_fibonacci()
    kesimpulan()
```
